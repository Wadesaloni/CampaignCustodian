```python
# ğŸ¤– CrewAI: Multi-Agent AI System Demo

## Comprehensive Guide to Building Autonomous AI Teams

This notebook demonstrates:
- **CrewAI Framework**: Building autonomous AI agents that work together
- **Agent Creation**: Defining specialized AI agents with specific roles and expertise
- **Task Management**: Creating structured tasks for agents to execute
- **Tool Integration**: Equipping agents with real-world tools and capabilities
- **Crew Execution**: Orchestrating multi-agent collaboration and workflow
- **Real-World Use Cases**: Market research, content creation, data analysis, and business intelligence

### ğŸ“š What is CrewAI?

CrewAI is a framework for building autonomous AI agent teams that can collaborate to solve complex tasks. Instead of using a single AI model, you create a "crew" of specialized agents that each have:
- **Specific Roles**: Defined expertise areas (Researcher, Writer, Analyst, etc.)
- **Clear Goals**: What each agent is trying to accomplish
- **Assigned Tools**: Functions and capabilities available to each agent
- **Backstory**: Context about the agent's expertise and personality

### ğŸ¯ Why Use Multi-Agent Systems?

- **Specialization**: Each agent focuses on what it does best
- **Parallel Processing**: Multiple agents can work on different tasks simultaneously
- **Quality**: Agents review and refine each other's work (like human collaboration)
- **Scalability**: Easy to add more agents for complex workflows
- **Transparency**: You can see what each agent is thinking and doing
- **Error Handling**: Failed tasks can be retried or delegated to other agents
```

## ğŸ“¦ Section 1: Initialize CrewAI Environment

First, we import all necessary libraries and set up API connections.

## ğŸ‘¥ Section 2: Create AI Agents with Specific Roles

Now we define multiple specialized agents that will form our crew. Each agent has:
- **Role**: Job title/specialization
- **Goal**: What they're trying to accomplish
- **Backstory**: Context about their expertise
- **Tools**: Functions they can use
- **LLM**: Which language model powers them

```python
# ============================================
# AGENT 1: Market Researcher
# ============================================
# This agent is an expert at finding market information

market_researcher = Agent(
    role="Market Research Analyst",  # Job title
    goal="""Conduct thorough market research and competitive analysis. 
    Find current market trends, customer needs, and competitive landscape.
    Provide data-driven insights for business decisions.""",  # What they aim to do
    backstory="""You are an experienced market analyst with 15 years in business intelligence.
    You excel at finding market gaps, understanding customer behavior, and identifying trends.
    You use data to support every conclusion and think like a strategic consultant.
    Your research helps companies make informed decisions about new markets and products.""",  # Their expertise
    llm=llm,  # Which LLM powers this agent
    verbose=True  # Print detailed reasoning
)

print("âœ… Market Researcher Agent Created")
print("   Role:", market_researcher.role)
print("   Goal: Conduct market analysis and competitive research")
```

```python
# ============================================
# AGENT 2: Business Strategist
# ============================================
# This agent develops strategic plans based on research

strategist = Agent(
    role="Business Strategy Consultant",  # Job title
    goal="""Develop comprehensive business strategies based on market insights.
    Create actionable plans that maximize opportunities and minimize risks.
    Think strategically about positioning, differentiation, and growth.""",  # Strategic goal
    backstory="""You are a senior business strategist who has helped 50+ companies succeed.
    You think 3-5 years ahead and identify strategic opportunities others miss.
    You combine market insights with operational excellence to create winning strategies.
    Your recommendations are known for being both visionary and practical.""",  # Expertise
    llm=llm,  # Same LLM
    verbose=True  # Show reasoning
)

print("âœ… Business Strategist Agent Created")
print("   Role:", strategist.role)
print("   Goal: Develop winning business strategies")
```

```python
# ============================================
# AGENT 3: Financial Analyst
# ============================================
# This agent handles financial planning and projections

financial_analyst = Agent(
    role="Financial Analyst",  # Job title
    goal="""Analyze financial viability and create realistic financial projections.
    Calculate ROI, identify cost drivers, and ensure financial sustainability.
    Provide detailed financial recommendations.""",  # Financial goal
    backstory="""You are a CFO-level financial analyst with expertise in startups and growth companies.
    You've managed P&Ls ranging from $1M to $100M+.
    You understand unit economics, cash flow management, and how to make money.
    Your financial models are known for accuracy and conservative estimates.""",  # Financial expertise
    llm=llm,  # Same LLM
    verbose=True  # Show reasoning
)

print("âœ… Financial Analyst Agent Created")
print("   Role:", financial_analyst.role)
print("   Goal: Create financial plans and projections")
```

```python
# ============================================
# AGENT 4: Crew Manager
# ============================================
# This agent synthesizes everyone's work into a cohesive plan

crew_manager = Agent(
    role="Chief Executive Officer",  # CEO role
    goal="""Synthesize research, strategy, and financial analysis into a comprehensive 
    business plan. Ensure alignment across all functions and create an executive summary.
    Make final recommendations to stakeholders.""",  # Management goal
    backstory="""You are an experienced CEO who has successfully launched multiple ventures.
    You excel at bringing teams together, synthesizing complex information, and making decisions.
    You think about the big picture and ensure all functions work in harmony.
    Your leadership style values data-driven decisions and clear communication.""",  # Leadership background
    llm=llm,  # Same LLM
    verbose=True  # Show reasoning
)

print("âœ… Crew Manager Agent Created")
print("   Role:", crew_manager.role)
print("   Goal: Synthesize all inputs into comprehensive strategy")

print("\n" + "="*70)
print("âœ… ALL AGENTS READY!")
print("="*70)
print("\nAgent Team:")
print(f"  1. {market_researcher.role}")
print(f"  2. {strategist.role}")
print(f"  3. {financial_analyst.role}")
print(f"  4. {crew_manager.role}")
```

## ğŸ“‹ Section 3: Define Tasks for Agent Collaboration

Tasks define WHAT work needs to be done. Each task is assigned to an agent and can depend on other tasks.
CrewAI orchestrates these tasks, passing outputs from one task as inputs to the next.

```python
# ============================================
# TASK 1: Market Research
# ============================================
# First task: Research the market

research_task = Task(
    # Description tells the agent what to do
    description="""Research a sustainable eco-friendly water bottle startup opportunity.
    
    Investigate:
    1. Current market size and growth rate for sustainable products
    2. Main competitors and their market positioning
    3. Customer demographics and pain points
    4. Current price points and market segments
    5. Regulatory requirements for manufacturing
    6. Distribution channels available (online, retail, wholesale)
    
    Provide specific data points, market insights, and competitive analysis.
    Format as structured bullet points with sources where possible.""",
    
    # Expected output format
    expected_output="""Comprehensive market research report with:
    - Market size and growth projections
    - 5+ key competitors with strengths/weaknesses
    - Target customer profiles
    - Pricing analysis
    - Distribution strategy recommendations
    - Key market opportunities and threats""",
    
    # Assign to market researcher
    agent=market_researcher
)

print("âœ… Research Task Created")
print("   Assigned to: Market Researcher")
print("   Focus: Eco-friendly water bottle market analysis")
```

```python
# ============================================
# TASK 2: Strategic Planning
# ============================================
# Second task: Develop strategy based on research

strategy_task = Task(
    description="""Develop a comprehensive business strategy for an eco-friendly water bottle startup.
    
    Based on the market research, create a strategy that includes:
    1. Unique value proposition and brand positioning
    2. Product differentiation strategy (materials, design, sustainability)
    3. Go-to-market strategy and launch plan
    4. Customer acquisition channels and marketing strategy
    5. Competitive advantages and defensibility
    6. Growth phases for years 1-3
    7. Team and organizational structure needed
    8. Key success metrics and KPIs
    
    Think strategically about how to stand out and win in this market.""",
    
    expected_output="""Strategic business plan including:
    - Clear positioning and unique value proposition
    - Detailed go-to-market strategy
    - Product development roadmap
    - Customer acquisition plan
    - 3-year growth strategy with milestones
    - Team structure and hiring plan
    - Key metrics to track success""",
    
    agent=strategist,
    # This task depends on research_task being completed first
    context=[research_task]  # Input from previous task
)

print("âœ… Strategy Task Created")
print("   Assigned to: Strategist")
print("   Depends on: Research Task (will use research output as input)")
```

```python
# ============================================
# TASK 3: Financial Projection
# ============================================
# Third task: Create financial projections

financial_task = Task(
    description="""Create detailed financial projections for the eco-friendly water bottle startup.
    
    Based on market research and strategy, develop:
    1. Startup costs and initial funding requirements
    2. Revenue projections (5 years) based on:
       - Unit pricing strategy
       - Sales volume assumptions
       - Market penetration rates
    3. Operating expense budget including:
       - Manufacturing/COGS
       - Marketing and customer acquisition costs
       - Personnel costs
       - Distribution and logistics
    4. Break-even analysis and payback period
    5. Cash flow projections (monthly for year 1, quarterly for years 2-3)
    6. Profitability timeline
    7. Capital requirements at each stage
    8. Key financial assumptions and sensitivities
    
    Be realistic and conservative with assumptions.""",
    
    expected_output="""Complete financial model including:
    - 5-year revenue projections
    - Detailed operating budget
    - Profitability analysis
    - Cash flow forecasts
    - Break-even calculations
    - Funding requirements
    - Key financial metrics (ROI, payback, margins)
    - Sensitivity analysis""",
    
    agent=financial_analyst,
    # This task depends on both research and strategy
    context=[research_task, strategy_task]  # Uses outputs from both previous tasks
)

print("âœ… Financial Task Created")
print("   Assigned to: Financial Analyst")
print("   Depends on: Research Task + Strategy Task")
```

```python
# ============================================
# TASK 4: Executive Summary
# ============================================
# Final task: Synthesize everything into an executive summary

executive_summary_task = Task(
    description="""Create a comprehensive executive summary and final business plan.
    
    Synthesize all research, strategy, and financial analysis into a cohesive plan.
    Create:
    1. Executive Summary (1-2 pages)
       - Business overview
       - Market opportunity
       - Solution/competitive advantage
       - Financial highlights
    2. Key recommendations for go/no-go decision
    3. Top 5 critical success factors
    4. Top 5 risks and mitigation strategies
    5. Recommended next steps and timeline
    6. Investment summary (if seeking funding)
    
    Make a clear recommendation: Should we pursue this opportunity? Why or why not?""",
    
    expected_output="""Final business plan document with:
    - Executive summary
    - Clear go/no-go recommendation
    - Critical success factors
    - Risk mitigation plan
    - Implementation roadmap
    - Key decision points and timing
    - Investment requirements
    - Expected outcomes and metrics""",
    
    agent=crew_manager,
    # This is the final task and depends on all previous tasks
    context=[research_task, strategy_task, financial_task]  # Uses all previous outputs
)

print("âœ… Executive Summary Task Created")
print("   Assigned to: CEO/Manager")
print("   Depends on: ALL previous tasks")
print("\n" + "="*70)
print("âœ… ALL TASKS DEFINED!")
print("="*70)
```

## ğŸš€ Section 4: Build and Execute Agent Crew

Now we assemble all agents and tasks into a Crew and execute the workflow.

```python
# ============================================
# CREATE THE CREW
# ============================================
# Assemble all agents and tasks into a single crew

startup_crew = Crew(
    # List of agents in the crew
    agents=[
        market_researcher,
        strategist,
        financial_analyst,
        crew_manager
    ],
    
    # List of tasks to execute
    tasks=[
        research_task,
        strategy_task,
        financial_task,
        executive_summary_task
    ],
    
    # Process type:
    # - Sequential: Tasks execute one after another (safer, linear)
    # - Hierarchical: Manager delegates to team members (more flexible)
    process=Process.sequential,  # Run tasks in order
    
    # Verbosity level (how much detail to print)
    verbose=True
)

print("âœ… Crew Assembled!")
print("\nğŸ“Š Crew Configuration:")
print(f"  â€¢ Agents: {len(startup_crew.agents)}")
print(f"  â€¢ Tasks: {len(startup_crew.tasks)}")
print(f"  â€¢ Process: Sequential (execute tasks in order)")
print(f"  â€¢ Verbose: Yes (show all reasoning)")
```

```python
# ============================================
# EXECUTE THE CREW
# ============================================
# This is where the magic happens!

print("\n" + "="*70)
print("ğŸš€ LAUNCHING STARTUP BUSINESS PLANNING CREW")
print("="*70)
print("\nâ±ï¸  Executing workflow...")
print("   â€¢ Step 1: Market Research")
print("   â€¢ Step 2: Strategic Planning")
print("   â€¢ Step 3: Financial Analysis")
print("   â€¢ Step 4: Executive Summary")
print("\nThis may take 2-5 minutes...\n")
print("="*70)

# Execute the crew - this runs all tasks in sequence
result = startup_crew.kickoff()

# Store the result for later use
final_business_plan = result
```

```python
# ============================================
# DISPLAY RESULTS
# ============================================
# Show the final output from the crew

print("\n" + "="*70)
print("âœ… CREW EXECUTION COMPLETE!")
print("="*70)

print("\nğŸ“‹ FINAL BUSINESS PLAN:")
print("-"*70)

# Display the result as formatted markdown
display(Markdown(f"""
## Complete Business Plan for Eco-Friendly Water Bottle Startup

{final_business_plan}

---

### Summary of Team Contributions:

**Market Researcher:** Conducted market analysis, identified competitors, and sized the opportunity

**Strategist:** Developed positioning, go-to-market strategy, and growth roadmap

**Financial Analyst:** Created financial projections and funding requirements

**CEO:** Synthesized all inputs into a comprehensive, actionable business plan
"""))
```

## ğŸ’¡ Section 5: Key Takeaways and Best Practices

### CrewAI Advantages:

1. **Specialization**: Each agent has a specific expertise and role
2. **Collaboration**: Agents build on each other's work
3. **Quality Control**: Different perspectives improve output quality
4. **Explainability**: You can see what each agent is thinking
5. **Scalability**: Easy to add more agents for complex problems

### When to Use Multi-Agent Systems:

âœ… **Use CrewAI when:**
- Task requires multiple perspectives and expertise areas
- You need to break down complex problems into subtasks
- Quality is more important than speed
- You want transparency in AI decision-making
- You're building intelligent automation workflows

âŒ **Don't use CrewAI when:**
- You need real-time, low-latency responses
- Task is simple and requires only one skill
- Cost optimization is the primary concern
- You need to integrate with strict APIs

### Best Practices:

1. **Clear Agent Definitions**: Define roles, goals, and backstories precisely
2. **Task Dependencies**: Sequence tasks logically, using outputs as inputs
3. **Tool Integration**: Give agents the right tools for their job
4. **Iterative Refinement**: Start simple, then add complexity
5. **Monitoring**: Track agent performance and refine prompts based on results
6. **Error Handling**: Design tasks with failure recovery in mind
