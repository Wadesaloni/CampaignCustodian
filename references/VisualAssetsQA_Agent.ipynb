```python
# Basic Visual Assets QA Agent
import os
import pandas as pd
from bs4 import BeautifulSoup

# Folders
inputs_folder = r"C:\Users\swade01\nxt24\InputsToQA"
reference_folder = r"C:\Users\swade01\nxt24\ReferenceFiles"

# Collect reference image filenames
reference_images = set()
for file in os.listdir(reference_folder):
    if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        reference_images.add(file)

print(f"Found {len(reference_images)} reference images.")

# Prepare report data
report = []

# Loop through all HTML files in InputsToQA
for file in os.listdir(inputs_folder):
    if file.lower().endswith('.html'):
        file_path = os.path.join(inputs_folder, file)
        with open(file_path, 'r', encoding='utf-8') as f:
            html_content = f.read()
        
        # Extract images
        soup = BeautifulSoup(html_content, "html.parser")
        images = [img.get('src') for img in soup.find_all('img') if img.get('src')]
        
        # Check each image against reference folder
        for img_src in images:
            img_name = os.path.basename(img_src)
            status = "PASS" if img_name in reference_images else "FAIL"
            report.append({
                "Email File": file,
                "Image Src": img_src,
                "Status": status
            })

# Save report
report_df = pd.DataFrame(report)
report_csv_path = os.path.join(inputs_folder, "VisualAssets_QA_Report.csv")
report_df.to_csv(report_csv_path, index=False)
print(f"Report saved at: {report_csv_path}")
```
